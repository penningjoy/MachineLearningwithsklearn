### Data Loading

Data is the main ingredient of any Machine Learning project. Loading data from varied sources is usually the first step in an ML Development Life Cycle.

For our purposes, we will using the built-in datasets mostly, for example, the iris.

```python
# to load the datasets
import sklearn.datasets as datasets

'''
A dataset is a dictionary like object containing data and related metadata. 

The two main components -- 
    data  - is an samples * features 2D Array.
    target - labels are stored here
'''
iris = datasets.load_iris()


# Get a view of how the data looks like 

print("The shape of data is {}" .format(iris.data.shape))
print("The shape of target is {} " .format(iris.target.shape))
print("Number of features in the IRIS dataset is {}" .format(iris.data.shape[1]))
print("Number of Samples - {} and Number of Labels - {}" .format(iris.data.shape[0], iris.target_names))
```

Scikit Learn can also help in generating dataset should you so want to. 

Let's try build the following datasets:

1.  Classification Dataset
2.  Regresstion Dataset

Generating Classification Dataset

```python
# Generating Classification Dataset

import numpy as np
import sklearn.datasets as datasets
import matplotlib.pyplot as plt


x, y = datasets.make_classification( n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1, random_state=45)

print("the shape of X dataset is {}" .format(x.shape))
print("the shape of Y dataset is {}" .format(y.shape))
print("")

# Plot them to visualise

%matplotlib inline
colours = np.random.rand(x.data.shape[0])
plt.scatter(x[:,0], x[:,1], c=colours)
plt.title("Classification Dataset generated by SKLEARN")
plt.show()
```

Generating Regression Dataset

```python
# Generating Regression Dataset

import numpy as np
import sklearn.datasets as datasets
import matplotlib.pyplot as plt


x, y = datasets.make_regression( n_features=1, n_informative=1)

print("the shape of X dataset is {}" .format(x.shape))
print("the shape of Y dataset is {}" .format(y.shape))
print("")
# Plot them to visualise

%matplotlib inline
colours = np.random.rand(x.data.shape[0])
plt.scatter(x, y, c=colours)
plt.title("Regression Dataset generated by SKLEARN")
plt.show()
```

#### Data Preprocessing

One of the most critical pieces in the Machine Learning lifecycle.  A lot of time is spent in cleaning, scaling, standardizing, normalizing the data before 
Machine Learning Algorithms can be applied on the data. Almost 70% of the time is spent in Data Preparation. 

### Data Scaling

Data Scaling is done to bring data in a similar range. 

i. MinMax Scaler

One of the scaling techniques. The formula is 

(max(x)−min(x)) / (​​x​i​​−min(x)​​)



```python
import sklearn.preprocessing as preprocessing
import numpy as np

X = np.random.randint(2, 10, size=(4, 2))  # creating a table ( Col - 2, Row - 4) with values ranging from 2 to 10
X2 = np.random.randint(100, 10000, size=(4, 2))  # creating a table ( Col - 2, Row - 4) with values ranging from 100 to 10000
X = np.concatenate((X, X2), axis=1)
print("Data without scaling")
print(X)


print("")

# Now applying MinMax Scaler
minmax = preprocessing.MinMaxScaler()
minmax.fit(X)
X_minmax = minmax.transform(X)
print("Data after applying Minmax")
print(X_minmax)
```

ii. Standardization

This one assumes that the distribution is normal. The formula is 

( x​i​​−mean(x)​​ ) / stdev(x)​​ 

```python
import sklearn.preprocessing as preprocessing
import numpy as np

X = np.random.randint(2, 10, size=(4, 2))  # creating a table ( Col - 2, Row - 4) with values ranging from 2 to 10
X2 = np.random.randint(100, 10000, size=(4, 2))  # creating a table ( Col - 2, Row - 4) with values ranging from 100 to 10000
X = np.concatenate((X, X2), axis=1)
print("Data without scaling")
print(X)

print("")

# Now applying Standardization 
std = preprocessing.MinMaxScaler()
std.fit(X)
X_std = std.transform(X)
print("Data after applying Standardization")
print(X_std)
```

iii. Binarizer

Binarizer function sets data to 1 or 0 based on a threshold. Say, a data value is above the threshold of 0.4, the binarizer will change it to 1 and 0 if otherwise.

```python
import sklearn.preprocessing as preprocessing
import numpy as np

X = np.random.randint(2, 10, size=(2, 5))  
binary = preprocessing.Binarizer(threshold= 4)
X_binarized = binary.transform(X)

print(" Original Data ")
print(" ------------- ")
print(X)

print(" ")

print(" Binarized Data ")
print(" -------------- ")
print(X_binarized)
```

iv. Label Enconder



```python
import sklearn.preprocessing as preprocessing
import numpy as np

X = np.array(["MotherTeresa","Wonderwoman", "Superman", "MKGandhi", "Ironman"])
labelencoder = preprocessing.LabelEncoder()
labelencoder.fit(X)
X_encoded = labelencoder.transform(X)


print(" Original Data ")
print(" ------------- ")
print(X)

print(" ")

print(" Label Encoded Data ")
print(" -------------- ")
print(X_encoded)
```
